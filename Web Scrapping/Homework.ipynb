```python
# Importing necessary libraries

import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
```

```python
# Membuat list kosong untuk menampung data

nama_produk = []
harga_produk = []
penjual = []
rating = []
terjual = []
kota_penjual = []

# Scrapping data dari Tokopedia untuk 10 halaman menggunakan for loop

for halaman in range(1,10):
    weblink = f'https://www.tokopedia.com/search?navsource=&page={halaman}&q=seblak&search_id=202509061228137A614FD06C325407ANZ6&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st='
    driver = webdriver.Chrome()
    driver.get (weblink)
    time.sleep(10)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # Mengambil data yang diinginkan menggunakan inspeksi elemen
    for kotak in soup.find_all('a', {'class': 'Ui5-B4CDAk4Cv-cjLm4o0g== XeGJAOdlJaxl4+UD3zEJLg=='}):
        
        # Mengambil data nama produk
        try:
            nama_produk.append(kotak.find('span', {'class': '+tnoqZhn89+NHUA43BpiJg=='}).get_text())
        except:
            nama_produk.append(None)

        # Mengambil data harga produk
        try:
            harga_produk.append(kotak.find('div', {'class': 'urMOIDHH7I0Iy1Dv2oFaNw=='}).get_text())
        except:
            harga_produk.append(None)

        # Mengambil data nama penjual
        try:
            penjual.append(kotak.find('span', {'class': 'si3CNdiG8AR0EaXvf6bFbQ== gxi+fsEljOjqhjSKqjE+sw== flip'}).get_text())
        except:
            penjual.append(None)

        # Mengambil rating produk
        try:
            rating.append(kotak.find('span', {'class': '_2NfJxPu4JC-55aCJ8bEsyw=='}).get_text())
        except:
            rating.append(None)

        # Mengambil data banyaknya produk terjual
        try:
            terjual.append(kotak.find('span', {'class': 'u6SfjDD2WiBlNW7zHmzRhQ=='}).get_text())
        except:
            terjual.append(None)

        # Mengambil kota penjual
        try:
            kota_penjual.append(kotak.find('span', {'class': 'gxi+fsEljOjqhjSKqjE+sw== flip'}).get_text())
        except:
            kota_penjual.append(None)
```

```python
# Melakukan pengecekan kepada data yang sudah diambil

harga_produk
```

```python
# Membuat DataFrame menggunakan pandas

df = pd.DataFrame({
    "nama_produk": nama_produk,
    "harga": harga_produk,
    "penjual": penjual,
    "rating": rating,
    "terjual": terjual,
    "kota_penjual": kota_penjual
})

print(df)

```

```python
# Mengubah format data harga menjadi angka dan mengganti nama variable menjadi harga_bersih

df["harga_bersih"] = (
    df["harga"]
    .str.replace("Rp", "", regex=False)   # hapus "Rp"
    .str.replace(".", "", regex=False)    # hapus titik pemisah ribuan
    .astype(int)                          # ubah ke integer
)


df.harga_bersih
```

```python
# Membersihkan data jumlah produk terjual

jumlah_terjual = []

for val in df["terjual"]:
    # Jika jumlah None, maka tetap akan None
    if pd.isna(val) :
        jumlah_terjual.append(None)
    else:
        # Jika tidak None, maka bersihkan teks menjadi integer/float krn ada nilai None
        val_clean = (
            str(val)
            .replace("rb", "") # menghapus rb
            .replace("+", "") # menghapus +
            .replace(" terjual", "") # menghapus terjual
            .strip() # membersihkan additional spasi yang mungkin masih ada
        )
        # Jika setelah di cleaning hasilnya kosong, maka akan diubah menjadi none
        if val_clean == "":
            jumlah_terjual_bersih.append(None)
        else:
            jumlah_terjual.append(int(val_clean))

# Memasukan hasil pembersihan ke dataframe
df["jumlah_terjual"] = jumlah_terjual

df


```

```python
# Menyimpan data ke csv

df.to_csv('tokopedia_seblak.csv', index = False)
```

```python
# Mengubah urutan dan isi data


new_order = ["nama_produk", "harga_bersih", "penjual", "kota_penjual", "rating", "jumlah_terjual"] # Urutan yang diinginkan

# Membuat df baru dan hanya menggunakan data
df_reorder = df.reindex(columns=new_order)

# Menyimpan data bersih ke csv
df_reorder.to_csv('tokopedia_seblak_clean.csv', index = False)

df_reorder
```
