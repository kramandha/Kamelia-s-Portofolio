## Soal 4: Hapus kolom show_id

```python
# Melakukan penghapusan kepada kolom show_id menggunakan del
del df['show_id']

df
```

## Soal 5: Ganti tipe data kolom date_added ke datetime64[ns]

```python
# Membersihkan space kosong pada date_added(karena awalnya ada error karena adanya additional space)
df['date_added'] = df['date_added'].str.strip()

# Mengubah struktur date_Added menjadi datetime64
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df
```

## Soal 6: Berapa banyak jumlah film (movies) dan TV Show pada data

```python
# Melakukan perhitungan values untuk kolom type
df['type'].value_counts()
```

## Soal 7: Film apa saja yang direlease di tahun 2021 yang berasal dari negara South Africa?

```python
# Melakukan 2 filter terhadap data, yaitu release_year == 2021 dan country == South Africa
df[(df['release_year'] == 2021) & (df['country'] == 'South Africa') & (df['type'] == "Movie")]
```

## Soal 8 = Tampilkan 10 negara menghasilkan film terbanyak all time (urutkan dari yang terbanyak hingga paling sedikit)!

```python
# Menampilkan data movie only dan menyusunnya berdasarkan dengan country. Lalu melakukan perhitungan kepada jumlah title yang ada dan mengurutkannya berdasarkan jumlah perhitungan title terbanyak dan menampilkan 10 data terbanyak
df[df['type'] == 'Movie'].groupby('country').agg(movie_count=('title','count')).sort_values('movie_count', ascending=False).head(10)
```

## Soal 9 = Ambil data film yang bergenre Comedies

```python
# Menyaring data listed_in yang memuat 'Comedies' dan type hanya 'Movies'
df[(df['listed_in'].str.contains('Comedies')) & (df['type'] == 'Movie')]
```

## Soal 10: Tampilkan film-film yang mengandung kata land di judulnya! (bisa gunakan method .str.contains)

```python
# Menyaring title yang mengandung 'land' dan menyaring type hanya 'Movie'
df[(df['title'].str.contains('land')) & (df['type'] == 'Movie')]
```

## Soal 11 = Siapa director yang paling banyak menghasilkan film dan TV Show?

```python
# Menyusun data berdasarkan director, menghitung jumlah title dan mengurutkan berdasarkan title terbanyak. Hanya menampilkan 1 nama teratas
df.groupby('director').agg(show_count=('title','count')).sort_values('show_count', ascending=False).head(1)
```

## Soal 12 = Negara mana saja yang menghasilkan TV Show terbanyak?

```python
# Memfilter type hanya TV Show dan mengkategorikannya berdasarkan country. Melakukan perhitungan jumlah title dan mengurutkannya secara descending. Hanya menampilkan 1 data teratas
df[df['type'] == 'TV Show'].groupby('country').agg(tvshow_count=('title','count')).sort_values('tvshow_count', ascending=False).head(1)
```
